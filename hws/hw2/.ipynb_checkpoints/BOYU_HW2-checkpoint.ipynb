```python
def predict(ham_prior, spam_prior, ham_like_dict, spam_like_dict, text):
    '''
    prediction function that uses prior and likelihood structure to compute proportional posterior for a single line of text
    '''
    #ham_spam_decision = 1 if classified as spam, 0 if classified as normal/ham
    ham_spam_decision = None




    '''YOUR CODE HERE'''
    #ham_posterior = posterior probability that the email is normal/ham
    ham_posterior = None

    #spam_posterior = posterior probability that the email is spam
    spam_posterior = None

    '''END'''
    given_spam_spam_total=1
    given_not_spam_notspam_total=1
    notinHamdic = min(ham_like_dict.values())
    notinSpamdic = min(spam_like_dict.values())
    content = text.split()
    
    for word in content:
        if word in ham_like_dict:
            given_not_spam_notspam_total = given_not_spam_notspam_total * ham_like_dict[word]
        else:
            given_not_spam_notspam_total = given_not_spam_notspam_total * notinHamdic
        if word in spam_like_dict:
            given_spam_spam_total = given_spam_spam_total * spam_like_dict[word]
        else:
            given_spam_spam_total = given_spam_spam_total * notinSpamdic

    ###############
    ###I ignored the denominator in both equation since they are the same!!!!!!!!!!!!!!

    ############
    ham_posteriror = ham_prior * given_not_spam_notspam_total
    spam_posteriror = spam_prior * given_spam_spam_total

    # 
    
    # spam_posteriror = Decimal(givenspamspamtotal/ ((givenspamspamtotal + givennotspamnotspamtotal)))
    # ham_posteriror = Decimal(givennotspamnotspamtotal/ ((givenspamspamtotal + givennotspamnotspamtotal)))
    
    if ham_posteriror < spam_posteriror:
        ham_spam_decision =1
    else:
        ham_spam_decision =0
    
    return ham_spam_decision
```

```python
def metrics(ham_prior, spam_prior, ham_dict, spam_dict, df):
    '''
    Calls "predict"
    '''
    hh = 0 #true negatives, truth = ham, predicted = ham
    hs = 0 #false positives, truth = ham, pred = spam
    sh = 0 #false negatives, truth = spam, pred = ham
    ss = 0 #true positives, truth = spam, pred = spam
    num_rows = df.shape[0]
    for i in range(num_rows):
        roi = df.iloc[i,:]
        roi_text = roi.text
        df['label_num'] = np.where(df["label"] == "ham", 0, 1)
        roi_label = roi.label_num
        guess = predict(ham_prior, spam_prior, ham_dict, spam_dict, roi_text)
        if roi_label == 0 and guess == 0:
            hh += 1
        elif roi_label == 0 and guess == 1:
            hs += 1
        elif roi_label == 1 and guess == 0:
            sh += 1
        elif roi_label == 1 and guess == 1:
            ss += 1
    
    acc = (ss + hh)/(ss+hh+sh+hs)
    precision = (ss)/(ss + hs)
    recall = (ss)/(ss + sh)
    return acc, precision, recall
    
if __name__ == "__main__":
    load("TEST_balanced_spam_ham.csv")
    '''YOUR CODE HERE'''
    #this cell is for your own testing of the functions above
```

```python

path_train = "./TRAIN_balanced_ham_spam.csv"
path_test = "./TEST_balanced_ham_spam.csv"
test_raw = load(path_test)
train_raw = load(path_train)


ham_p_train, spam_p_train = prior(train_raw)
ham_dict_train, spam_dict_train = likelihood(train_raw)
train_acc, train_precision, train_recall = metrics(ham_p_train, spam_p_train, ham_dict_train, spam_dict_train, train_raw)


ham_p_test, spam_p_test = prior(test_raw)
ham_dict_test, spam_dict_test = likelihood(test_raw)
test_acc, test_precision, test_recall  = metrics(ham_p_test, spam_p_test, ham_dict_test, spam_dict_test, test_raw)

print("train",train_acc, train_precision, train_recall)
print("test",test_acc, test_precision, test_recall)
```
